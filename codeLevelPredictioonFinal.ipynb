```python
from contextlib import asynccontextmanager
import ngrok
import uvicorn
from fastapi import FastAPI
from pydantic import BaseModel
```

```python
NGROK_AUTH_TOKEN="30rExtbRV1KoCCqQs7LpRQxmy0f_2KWjjMZgJ3D4X1n1X8F5y"
APPLICATION_PORT=8000
```

```python
@asynccontextmanager
async def lifespan(app: FastAPI):
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    ngrok.forward(
        addr=APPLICATION_PORT,
        domain="redfish-prime-bluejay.ngrok-free.app"
    )
    yield
    ngrok.disconnect()


import nest_asyncio
nest_asyncio.apply()
app = FastAPI(lifespan=lifespan)


class Prompt(BaseModel):
    text: str

@app.post("/generate/")
def generate_text(prompt: Prompt):
    template = """Analyze the following code function and classify them based on the difficulty level into beginner, intermediate and advanced. Output only beginner, intermediate or advanced:

    Code:
    {code}

    """
    sample_prompt = template.format(code=prompt.text)
    output = model.generate(sample_prompt)
    return {"response": output.replace(f"{sample_prompt}", ""), "fullResponse": output}
```

```python
uvicorn.run(app, host="127.0.0.1", port=APPLICATION_PORT)
```
